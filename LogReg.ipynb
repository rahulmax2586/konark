```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import datasets , metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
data = datasets.load_digits()
```

```python
plt.figure(figsize=(20,4))
for index , (image , label) in enumerate(zip(data.data[0:5] , data.target[0:5])):
    plt.subplot(1,5,index + 1)
    plt.imshow( np.reshape(image , (8 , 8)) , cmap=None )
```

```python
X_train , X_test , y_train , y_test = train_test_split( data.data , data.target , test_size = 0.5 )
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train , y_train)
```

```python
model.predict(X_test[0:5])
```

```python
model.score(X_test , y_test)
```

```python
prediction = model.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_test , prediction)
```

```python
cm
```

```python
plt.figure(figsize=(10 , 10))
sns.heatmap(cm , cmap=plt.cm.afmhot_r , square=True , linecolor='red' , linewidths=0.5 , fmt="0.2f" , annot=True )
```
